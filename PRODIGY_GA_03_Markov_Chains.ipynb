```python
with open("Markov_Chains_Dataset.txt", "r", encoding="utf-8") as file:
    text = file.read()

print(text[:1000])
```

```python
import markovify

model = markovify.Text(text)
```

```python
print(model.make_sentence())
```

```python
for i in range(5):
    print(f"Sentence {i+1}:")
    print(model.make_sentence())
    print()
```

```python
for i in range(5):
    print(model.make_short_sentence(120))
    print()
```

```python
print(model.make_sentence_with_start("Artificial"))
```

```python
print(model.make_sentence_with_start("Cybersecurity"))
```

```python
print("Generated Text using Markov Chains\n")

count = 1

while count <= 10:
    sentence = model.make_sentence()

    if sentence:
        print(f"{count}. {sentence}\n")
        count += 1
```
